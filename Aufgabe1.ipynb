```python
from torchvision import models
import matplotlib.pyplot as plt
from torchvision import transforms
from PIL import Image
import torch
import numpy as np

# laden eines 101-schichtigem neuronalem Faltungsnetzwerks:
resnet = models.resnet101(pretrained=True) 
alexnet = models.AlexNet()

# Eingabebilder vorverarbeiten (selbe Größe, ähnliche Farbwerte etc.):
preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )])

# Öffnen der Katzenbilder:
img = Image.open("Katze.jpg")
img1 = Image.open("Katze1.jpg")
img2 = Image.open("Katze2.jpg")
img3 = Image.open("Katze3.jpg")
img4 = Image.open("Katze4.jpg")
img5 = Image.open("Katze5.jpg")
img6 = Image.open("Katze6.jpg")
img7 = Image.open("Katze7.jpg")
img8 = Image.open("Katze8.jpg")
img9 = Image.open("Katze9.jpg")

image_datas = [img,img1,img2,img3,img4,img5,img6,img7,img8,img9]

# Vorverarbeitung der Katzenbilder:
img_t = preprocess(img); img_t1 = preprocess(img1); img_t2 = preprocess(img2); img_t3 = preprocess(img3); img_t4 = preprocess(img4); img_t5 = preprocess(img5); img_t6 = preprocess(img6); img_t7 = preprocess(img7); img_t8 = preprocess(img8); img_t9 = preprocess(img9)
batch_t = torch.unsqueeze(img_t, 0); batch_t1 = torch.unsqueeze(img_t1, 0); batch_t2 = torch.unsqueeze(img_t2, 0); batch_t3 = torch.unsqueeze(img_t3, 0); batch_t4 = torch.unsqueeze(img_t4, 0); batch_t5 = torch.unsqueeze(img_t5, 0); batch_t6 = torch.unsqueeze(img_t6, 0); batch_t7 = torch.unsqueeze(img_t7, 0); batch_t8 = torch.unsqueeze(img_t8, 0); batch_t9 = torch.unsqueeze(img_t9, 0)

# Anwendung resnet/alexnet:
out = resnet(batch_t); out1 = resnet(batch_t1); out2 = resnet(batch_t2); out3 = resnet(batch_t3); out4 = resnet(batch_t4); out5 = resnet(batch_t5); out6 = resnet(batch_t6); out7 = resnet(batch_t7); out8 = resnet(batch_t8); out9 = resnet(batch_t9)
outa = alexnet(batch_t); out1a = alexnet(batch_t1); out2a = alexnet(batch_t2); out3a = alexnet(batch_t3); out4a = alexnet(batch_t4); out5a = alexnet(batch_t5); out6a = alexnet(batch_t6); out7a = alexnet(batch_t7); out8a = alexnet(batch_t8); out9a = alexnet(batch_t9)

with open('imagenet_classes.txt') as f: labels = [line.strip() for line in f.readlines()]
_, index = torch.max(out, 1); _, index1 = torch.max(out1, 1); _, index2 = torch.max(out2, 1); _, index3 = torch.max(out3, 1); _, index4 = torch.max(out4, 1); _, index5 = torch.max(out5, 1); _, index6 = torch.max(out6, 1); _, index7 = torch.max(out7, 1); _, index8 = torch.max(out8, 1); _, index9 = torch.max(out9, 1)
_, indexa = torch.max(outa, 1); _, index1a = torch.max(out1a, 1); _, index2a = torch.max(out2a, 1); _, index3a = torch.max(out3a, 1); _, index4a = torch.max(out4a, 1); _, index5a = torch.max(out5a, 1); _, index6a = torch.max(out6a, 1); _, index7a = torch.max(out7a, 1); _, index8a = torch.max(out8a, 1); _, index9a = torch.max(out9a, 1)

# Berechnung der Prozentsätze mit resnet/alexnet: 
percentage = torch.nn.functional.softmax(out, dim=1)[0] * 100; percentage1 = torch.nn.functional.softmax(out1, dim=1)[0] * 100; percentage2 = torch.nn.functional.softmax(out2, dim=1)[0] * 100; percentage3 = torch.nn.functional.softmax(out3, dim=1)[0] * 100; percentage4 = torch.nn.functional.softmax(out4, dim=1)[0] * 100; percentage5 = torch.nn.functional.softmax(out5, dim=1)[0] * 100; percentage6 = torch.nn.functional.softmax(out6, dim=1)[0] * 100; percentage7 = torch.nn.functional.softmax(out7, dim=1)[0] * 100; percentage8 = torch.nn.functional.softmax(out8, dim=1)[0] * 100; percentage9 = torch.nn.functional.softmax(out9, dim=1)[0] * 100
percentagea = torch.nn.functional.softmax(outa, dim=1)[0] * 100; percentage1a = torch.nn.functional.softmax(out1a, dim=1)[0] * 100; percentage2a = torch.nn.functional.softmax(out2a, dim=1)[0] * 100; percentage3a = torch.nn.functional.softmax(out3a, dim=1)[0] * 100; percentage4a = torch.nn.functional.softmax(out4a, dim=1)[0] * 100; percentage5a = torch.nn.functional.softmax(out5a, dim=1)[0] * 100; percentage6a = torch.nn.functional.softmax(out6a, dim=1)[0] * 100; percentage7a = torch.nn.functional.softmax(out7a, dim=1)[0] * 100; percentage8a = torch.nn.functional.softmax(out8a, dim=1)[0] * 100; percentage9a = torch.nn.functional.softmax(out9a, dim=1)[0] * 100

p = labels[index[0]], percentage[index[0]].item(); p1 = labels[index[0]], percentage1[index[0]].item(); p2 = labels[index[0]], percentage2[index[0]].item(); p3 = labels[index[0]], percentage3[index[0]].item(); p4 = labels[index[0]], percentage4[index[0]].item(); p5 = labels[index[0]], percentage5[index[0]].item(); p6 = labels[index[0]], percentage6[index[0]].item(); p7 = labels[index[0]], percentage7[index[0]].item(); p8 = labels[index[0]], percentage8[index[0]].item(); p9 = labels[index[0]], percentage9[index[0]].item()
pa = labels[indexa[0]], percentagea[indexa[0]].item(); p1a = labels[indexa[0]], percentage1a[indexa[0]].item(); p2a = labels[indexa[0]], percentage2a[indexa[0]].item(); p3a = labels[indexa[0]], percentage3a[indexa[0]].item(); p4a = labels[indexa[0]], percentage4a[indexa[0]].item(); p5a = labels[indexa[0]], percentage5a[indexa[0]].item(); p6a = labels[indexa[0]], percentage6a[indexa[0]].item(); p7a = labels[indexa[0]], percentage7a[indexa[0]].item(); p8a = labels[indexa[0]], percentage8a[indexa[0]].item(); p9a = labels[indexa[0]], percentage9a[indexa[0]].item() 

f, ax = plt.subplots(5,2)
f.set_figheight(8) 
f.set_figwidth(8)

ax[0,0].imshow(image_datas[0]); ax[0,0].set_title('resnet:'+str(p)+ "\n" +'alexnet:'+str(pa))
ax[1,0].imshow(image_datas[1]); ax[1,0].set_title('resnet:'+str(p1)+ "\n" +'alexnet:'+str(p1a)) 
ax[2,0].imshow(image_datas[2]); ax[2,0].set_title('resnet:'+str(p2)+ "\n" +'alexnet:'+str(p2a)) 
ax[3,0].imshow(image_datas[3]); ax[3,0].set_title('resnet:'+str(p3)+ "\n" +'alexnet:'+str(p3a)) 
ax[4,0].imshow(image_datas[4]); ax[4,0].set_title('resnet:'+str(p4)+ "\n" +'alexnet:'+str(p4a)) 

ax[0,1].imshow(image_datas[5]); ax[0,1].set_title('resnet:'+str(p5)+ "\n" +'alexnet:'+str(p5a)) 
ax[1,1].imshow(image_datas[6]); ax[1,1].set_title('resnet:'+str(p6)+ "\n" +'alexnet:'+str(p6a)) 
ax[2,1].imshow(image_datas[7]); ax[2,1].set_title('resnet:'+str(p7)+ "\n" +'alexnet:'+str(p7a)) 
ax[3,1].imshow(image_datas[8]); ax[3,1].set_title('resnet:'+str(p8)+ "\n" +'alexnet:'+str(p8a)) 
ax[4,1].imshow(image_datas[9]); ax[4,1].set_title('resnet:'+str(p9)+ "\n" +'alexnet:'+str(p9a)) 


plt.subplots_adjust(wspace=6, 
                    hspace=3)
```
